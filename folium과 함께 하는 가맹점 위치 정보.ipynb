위도경도 변환을 통해 각 가맹점의 위치를 정확하게 파악할 수 있게 되었다.<br>
주소를 얻고 보니, 어느 구에 얼마만큼의 가맹점이 있는지 지도 위에 시각적으로 표현하고 싶어졌다.<br>

```python
import pandas as pd
import json
```

```python
df=pd.read_excel('가맹점 정보_데이터베이스 연결용.xlsx', sheet_name="Sheet1")
df2 = df.iloc[0:20]
```

```python
# 가장 먼저. 지도에 표시하기 위해서는 두가지 방법 : 1)주소로 표시 2)위도경도로 표시 이 존재한다.
# 주소의 경우 현재 도로명/주소명이 섞여 있으므로, 일단 위도경도로 먼저 도전
stores = df2.loc[:,["store_name","longitude","Latitude"]]
```

```python
import folium
# 지도 시각화를 위해 conda에서 folium을 설치한 후 import해준다
```

```python
map_osm = folium.Map(location=[35.200014293506754, 129.05446167208507],  tiles='stamentoner', zoom_start=11)
# 시작지점(location)과 확대 정도를 설정해준 후 지도를 생성하자
```

```python
marker1 = folium.Marker([35.19979645787612, 129.05675238630965], 
                        popup="house", 
                        icon=folium.Icon(color='red'))
# 생성하고자 하는 위치의 위도경도, 클릭 시 팝업, 아이콘 을 설정한 후 marker를 생성한다
marker1.add_to(map_osm)  
# 마커 객체를 지도에 추가
map_osm
```

```python
for i in stores.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = stores.loc[i, 'store_name']
    lat = stores.loc[i, 'Latitude']
    lng = stores.loc[i, 'longitude']
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], popup=name)
    marker.add_to(map_osm)

map_osm
```
